```python
#import bin-encoding data
import pandas as pd
import numpy as np
import scipy.stats as stats
import sklearn
import random
import os
from pathlib import Path
from sklearn.linear_model import *
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def LoadData(filename):
    DATA_DIR = "Data"
    ENCODING_DIR = os.path.join(DATA_DIR, filename)
    data = pd.read_csv(ENCODING_DIR, index_col=[0])
    return data
```

```python
filename = "binEncoding.csv"
```

```python
X = LoadData(filename)
Temp = X.drop(columns=['article_id'])
X.head()
```

```python
#Similarity matrix between each article
Csim = cosine_similarity(Temp)

#convert to numpy
npV = np.asarray(Csim)
print(npV)
rows = np.size(npV,0)
print(rows)

#print(Csim)
#print(len(Csim))

```

```python
Y = X
#Y['relatedArticles'] = 0
#Y['SecondRelated'] = 0
#Y['ThirdRelated'] = 0
Y.head(10)
```

```python
for i in range(rows):
    #find most related articles indexed
    a = npV[i,:]
    index = np.argpartition(a, -4)[-4:]
    index2= index[np.argsort(a[index])]
    
    #show the index in X matrix
    print(i)
    print(index)
    print(index2)
    #show the similarity value
    print(a[index2])
    related = []
    related.append(str(X.iloc[index2[2]]['article_id']))
    related.append(str(X.iloc[index2[1]]['article_id']))
    related.append(str(X.iloc[index2[0]]['article_id']))
    
    
    Y.at[i, 'related_articles'] = ', '.join(related)
    
    Final = Y[['article_id', 'related_articles']]
    #Put into table Y
    #Y.at[i, 'FirstRelated'] = FirstID
    #Y.at[i, 'SecondRelated'] = SecondID
    #Y.at[i,'ThirdRelated'] = ThirdID
    
```
